```python
class DecisionTreeRegressor:
    def __init__(self,max_depth=8,min_samples_split=3):#Trial and Error
        self.max_depth=max_depth
        self.min_samples_split=min_samples_split
        self.tree=None
    def fit(self,X,y):
        data=np.concatenate([X,y.reshape(-1,1)],axis=1)
        self.tree=self.build_tree(data)
    def predict_single(self,sample,tree):
        if "value" in tree:
            return tree["value"]
        feature_value=sample[tree["feature_index"]]
        if feature_value<tree["threshold"]:
            return self.predict_single(sample,tree["left"])
        else:
            return self.predict_single(sample,tree["right"])
    def predict(self,X):
        return np.array([self.predict_single(sample,self.tree) for sample in X])
    def build_tree(self,data,depth=0):
        X,y=data[:,:-1],data[:,-1]
        num_samples,num_features=X.shape
        if(self.max_depth<=depth or num_samples<self.min_samples_split):
            return {"value":np.mean(y)}
        best_split=self.find_best_split(data,num_features)
        if not best_split:
            return {"value":np.mean(y)}
        left_tree=self.build_tree(best_split["left_data"],depth+1)
        right_tree=self.build_tree(best_split["right_data"],depth+1)
        return {"feature_index":best_split["feature_index"],
                "threshold":best_split["threshold"],
                 "left":left_tree,
                 "right":right_tree,
               }
    def find_best_split(self,data,num_features):
        X,y=data[:,:-1],data[:,-1]
        best_mse=float("inf")
        best_split=None
        for feature_index in range(num_features):
            thresholds=np.unique(X[:,feature_index])
            for threshold in thresholds:
                left_data,right_data=self.split(data,feature_index,threshold)
                if(len(left_data)==0 or len(right_data)==0):
                    continue
                mse=self.calculate_mse(left_data[:,-1],right_data[:,-1])
                if mse<best_mse:
                    best_split={"feature_index":feature_index,
                                "threshold":threshold,
                                "left_data":left_data,
                                "right_data":right_data
                               }
                    best_mse=mse
        return best_split
    def split(self,data,feature_index,threshold):
        left_mask=data[:,feature_index]<threshold
        right_mask=data[:,feature_index]>=threshold
        return data[left_mask],data[right_mask]
    def calculate_mse(self,left_y,right_y):
        left_mean=np.mean(left_y)
        right_mean=np.mean(right_y)
        left_mse=np.sum((left_y-np.mean(left_y))**2)/len(left_y)
        right_mse=np.sum((right_y-np.mean(right_y))**2)/len(right_y)
        total_mse=((len(left_y)*left_mse)+(len(right_y)*right_mse))/(len(left_y)+len(right_y))
        return total_mse
```

```python
class k_folds_cross_val:
    def __init__(self,k_folds=5):
        self.k_folds=k_folds
    def cross_val_score(self,data):
        mse_list=list()
        model=DecisionTreeRegressor()
        for i in range(self.k_folds):
            train,test=self.train_test_split(data,i+1)
            X_train,y_train=train[:,:-1],train[:,-1]
            model.fit(X_train,y_train)
            X_test,y_test=test[:,:-1],test[:,-1]
            predictions=model.predict(X_test)
            mse=(np.sum((y_test-predictions)**2))/len(predictions)
            print(r2_score(y_test,predictions))
            mse_list.append(mse)
        return mse_list
    def train_test_split(self,data,iteration):
        m,n=data.shape
        t=int(m/self.k_folds)
        train=np.concatenate([data[iteration*t:,:],data[0:(iteration-1)*t,:]],axis=0)
        test=data[(iteration-1)*t:iteration*t,:]
        return train,test
```

```python
def r2_score(true_value,predictions):
    true_value=true_value.reshape(1,-1)
    predictions=predictions.reshape(1,-1)
    mean=np.mean(true_value)
    r2=1-(np.sum((true_value-predictions)**2))/(np.sum((true_value-mean)**2))
    return r2
```

```python
np.random.seed(72)
per=np.random.permutation(2500)
data_ndarray=data_set.to_numpy()
data_shuffled=data_ndarray[per]
cross_validator=k_folds_cross_val()
score=cross_validator.cross_val_score(data_shuffled)
print(np.mean(score),np.std(score))
```
