```python
import torch
from torch.utils.data import DataLoader
from torchvision import datasets,transforms
torch.manual_seed(420)
import os
import zipfile
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score,ConfusionMatrixDisplay
import seaborn as sns
```

<h1>Data Loading and Preprocessing</h1>

```python
with zipfile.ZipFile('A,B,CNNS_with_Tim.zip') as file:
    file.extractall()
train_path='A,B,CNNS_with_Tim/veggie_heap_training'
test_path='A,B,CNNS_with_Tim/veggie_heap_testing'
train_transformer=transforms.Compose([
    transforms.TrivialAugmentWide(),
    transforms.Resize((256,256)),
    transforms.ToTensor()
])
test_transformer=transforms.Compose([transforms.Resize((256,256)),transforms.ToTensor()])
train_set=datasets.ImageFolder(root=train_path,transform=train_transformer,target_transform=None)
test_set=datasets.ImageFolder(root=test_path,transform=test_transformer,target_transform=None)
display(train_set,test_set)
```

```python
train_dataloader=DataLoader(dataset=train_set,
                           batch_size=100,
                           num_workers=12,
                           shuffle=True,
                           pin_memory=True)
test_dataloader=DataLoader(dataset=test_set,
                          batch_size=100,
                          num_workers=12,
                          shuffle=True,
                          pin_memory=True)
display(train_dataloader,test_dataloader)
```

```python
plt.figure(figsize=(12,9))
for i in range(9):
    plt.subplot(3,3,i+1)
    img,label=next(iter(train_dataloader))
    plt.imshow(img[0].permute(2,1,0))
    plt.title(train_set.classes[label[0]])
    plt.axis("Off")
plt.show()
```

<h1>Model Development and Training</h1>

```python
class CNN(torch.nn.Module):
    def __init__(self,num_classes,in_channels=3):
        super(CNN,self).__init__()
        self.layer1=torch.nn.Sequential(
            torch.nn.Conv2d(3,32,kernel_size=3,stride=1,padding=1),
            torch.nn.ReLU(),
            torch.nn.MaxPool2d(kernel_size=2,stride=2))
        self.layer2=torch.nn.Sequential(
            torch.nn.Conv2d(32,64,kernel_size=3,stride=1,padding=1),
            torch.nn.ReLU(),
            torch.nn.MaxPool2d(kernel_size=2,stride=2))
        self.layer3=torch.nn.Sequential(
            torch.nn.Conv2d(64,128,kernel_size=3,stride=1,padding=1),
            torch.nn.ReLU(),
            torch.nn.MaxPool2d(kernel_size=2,stride=2))
        self.fc=torch.nn.Linear(32*32*128,num_classes,bias=True)
        torch.nn.init.xavier_uniform(self.fc.weight)
    def forward(self,X):
        out=self.layer1(X)
        out=self.layer2(out)
        out=self.layer3(out)
        out=out.view(out.size(0),-1)
        out=self.fc(out)
        return out
```

```python
device=torch.device('cuda' if torch.cuda.is_available() else ('mps' if torch.backends.mps.is_available() else'cpu'))
num_classes=len(train_set.classes)
learning_rate=0.001
batch_size=100
num_epochs=200
```

```python
model=CNN(num_classes).to(device)
#model.load_state_dict(torch.load('model.pth'))
```

```python
criterion=torch.nn.CrossEntropyLoss().to(device)
optimizer=torch.optim.Adam(model.parameters(),learning_rate)

# total_batch=len()
loss=dict()

for epoch in range(num_epochs):
    batch_cost=0
    for X,y in train_dataloader:
        X=X.to(device)
        y=y.to(device)
        optimizer.zero_grad()
        hypothesis=model(X)
        cost=criterion(hypothesis,y)
        cost.backward()
        optimizer.step()
        batch_cost+=cost
        del X
        del y
        torch.cuda.empty_cache()
    batch_cost/=128
    loss[epoch]=batch_cost
    print(batch_cost)
```

```python
torch.save(model.state_dict(),'model')
```

<h1>Model Evaluation</h1>

```python
test_dataloader=DataLoader(dataset=test_set,
                          batch_size=10,
                          num_workers=12,
                          shuffle=True,
                          pin_memory=True)
import numpy as np
predictions=[]
actual=[]
for X,y in test_dataloader:
    X=X.to(device)
    y=y.to(device)
    hypothesis=model(X)
    _,predicted_class=torch.max(hypothesis,dim=1)
    predictions.append(predicted_class.cpu().numpy())
    actual.append(y.cpu().numpy())
    del X
    del y
```

```python
predictions_flat = np.concatenate(predictions)
actual_flat = np.concatenate(actual)
print("Accuracy",np.mean(actual_flat==predictions_flat))
```

```python
epochs=loss.keys()
loss_values=[loss[key].item() for key in epochs]
plt.figure(figsize=(8,8))
plt.plot(epochs,loss_values)
plt.xlabel("Number of Epochs")
plt.ylabel("Cross Entropy Loss")
plt.title("Loss over epochs")
plt.text(140,1.43,"Batch size of 100 was used")
plt.show()
```

```python
plt.figure(figsize=(10,10))
corr_matr=pd.DataFrame(confusion_matrix(actual_flat,predictions_flat),index=train_set.classes,columns=train_set.classes)
sns.heatmap(corr_matr,lw=0.1,fmt='d',cmap='Blues',annot=True)
plt.title("Confusion Matrix for CNN Classifier")
plt.xticks(rotation=45)
plt.show()
```

```python
# a=accuracy_score(actual_flat,predictions_flat)
p=precision_score(actual_flat,predictions_flat,average=None)
r=recall_score(actual_flat,predictions_flat,average=None)
f1=f1_score(actual_flat,predictions_flat,average=None)
metrics_table=pd.DataFrame({'Precision':p,'Recall':r,'F1-Score':f1},index=train_set.classes)

plt.figure(figsize=(9,7))
sns.heatmap(metrics_table,fmt='.2f',annot=True,lw=0.01,cmap='YlGnBu',cbar=True)
plt.title("Precision, Recall and F1-Score by Class")
plt.yticks(rotation=0)
plt.show()
```
