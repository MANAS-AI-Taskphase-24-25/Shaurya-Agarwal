```python
def train_test_split(data,split_ratio):
    m,n=data.shape
    index=np.random.permutation(m)
    train=data[index[:int(m*(1-split_ratio))]]
    test=data[index[int(m*split_ratio):]]
    return train,test
train,test=train_test_split(data.to_numpy(),0.2)
def normalise(data):
    return (data-np.mean(data,axis=0))/np.std(data,axis=0)
train_normalised=normalise(train[:,:-1])
test_normalised=normalise(test[:,:-1])
```

```python
def silhouette_score(X,labels):
    m,n=X.shape
    distances=np.sqrt(np.sum((X[:,np.newaxis,:]-X[np.newaxis,:,:])**2,axis=2))
    unique_labels=np.unique(labels)
    n_clusters=len(unique_labels)
    if n_clusters==1:
        raise ValueError("Silhouette score cannot be computed for a single cluster")
    silhouette_scores=np.zeros(m)
    for i in range(m):
        current_label=labels[i]
        intra_cluster_distances=distances[i][labels==current_label]
        a_i=np.mean(intra_cluster_distances[intra_cluster_distances!=0])
        b_i=np.inf
        for j in unique_labels[unique_labels!=current_label]:
            inter_cluster_distances=distances[i][labels==j]
            b_i=np.min([np.mean(distances[i][labels==j]) for j in unique_labels[unique_labels!=current_label]])
        silhouette_scores[i]=(b_i-a_i)/np.max([a_i,b_i])
    score=np.mean(silhouette_scores)
    fig,ax=plt.subplots(figsize=(10,6))
    y_lower=10
    for i in unique_labels:
        cluster_silhouette_score=silhouette_scores[labels==i]
        cluster_silhouette_score.sort()
        y_upper=y_lower+len(cluster_silhouette_score)
        ax.fill_betweenx(np.arange(y_lower,y_upper),0,cluster_silhouette_score,alpha=0.5)
        ax.text(-0.05,y_lower+0.5*len(cluster_silhouette_score),str(i))
        y_lower=y_upper+10
    ax.axvline(score,color="red",linestyle="--")
    ax.set_title("Silhouette Diagram",fontsize=14)
    ax.set_xlabel("Silhouette Coefficients",fontsize=12)
    ax.set_ylabel("Clusters",fontsize=12)
    ax.set_yticks([])
    ax.set_xlim([-0.1,1])
    plt.show()
    return score
```

```python
class KMeans:
    def __init__(self,n_clusters,cluster_centers_=None,n_iter=10000):
        self.n_clusters=n_clusters
        self.cluster_centers=cluster_centers_
        self.n_iter=n_iter
    def fit(self,X):
        m,n=X.shape
        if self.cluster_centers==None:
            self.cluster_centers=X[np.random.choice(m,self.n_clusters,replace=False)]
        distort_old=np.inf
        distort_new=-np.inf
        iterations=0
        while(distort_old>=distort_new and iterations<self.n_iter):
            distances=np.linalg.norm(X[:,np.newaxis,:]-self.cluster_centers[np.newaxis,:,:],axis=2)
            clusters=np.argmin(distances,axis=1)
            self.cluster_centers=np.array([np.mean(X[clusters==j],axis=0) for j in range(len(self.cluster_centers))])
            distort_old=distort_new
            distort_new=self.distortion(X,clusters)
            iterations+=1
        return clusters
    def predict(self,X):
        distances=np.linalg.norm(X[:,np.newaxis,:]-self.cluster_centers[np.newaxis,:,:],axis=2)
        clusters=np.argmin(distances,axis=1)
        return clusters
    def distortion(self,X,y):
        return np.sum(np.sum((X-self.cluster_centers[y])**2,axis=1))
    def inertia(self,X,y):
        return np.mean(np.sum((X-self.cluster_centers[y])**2,axis=1))
```

```python
best_accu=-np.inf
for i in range(10000):
    kmeans=KMeans(3)
    y=kmeans.fit(train_normalised)
    train_cluster_labels=kmeans.predict(train_normalised)+1
    accuracy=np.mean(train_cluster_labels==train[:,-1])
    if accuracy>=best_accu:
        best_accu=accuracy
        best_model=kmeans
final_model=best_model
test_cluster_labels=final_model.predict(test_normalised)+1
final_accuracy=np.mean(test_cluster_labels==test[:,-1])
final_accuracy
```

```python
class PCA:
    def __init__(self,final_no_of_dimensions=2):
        self.final_no_of_dimensions=final_no_of_dimensions
    def fit(self,X):
        U, s, Vt = np.linalg.svd(X)
        self.Wd=Vt.T[:,:self.final_no_of_dimensions]
    def transform(self,X):
        return X@self.Wd
    def fit_transform(self,X):
        self.fit(X)
        return self.transform(X)
pca=PCA(3)
X_transformed=pca.fit_transform(train_normalised)
cluster_centers_transformed=pca.transform(final_model.cluster_centers)
fig=plt.figure(figsize=(50,50))
ax1=fig.add_subplot(111,projection='3d')
ax1.scatter(*X_transformed.T,c=final_model.predict(train_normalised),cmap='viridis',s=100,vmin=-1,vmax=3)
ax1.scatter(*cluster_centers_transformed.T,marker='X',color='red',s=400,label='Cluster Centers',edgecolors="k")
ax1.set_title("3-D Projection",fontsize=80)
ax1.legend(fontsize=40)
plt.show()
```

```python
pca=PCA()
X_transformed=pca.fit_transform(train_normalised)
cluster_centers_transformed=pca.transform(final_model.cluster_centers)
fig=plt.figure(figsize=(14,7))
ax1=fig.add_subplot(111)
ax1.scatter(*X_transformed.T,c=final_model.predict(train_normalised),cmap='viridis',s=10)
ax1.scatter(*cluster_centers_transformed.T,marker='X',color='red',s=40,label='Cluster Centers',edgecolors="k")
ax1.set_title("2-D Projection")
ax1.legend()
plt.grid()
plt.show()
```
